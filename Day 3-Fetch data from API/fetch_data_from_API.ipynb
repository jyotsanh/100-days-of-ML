```python
data = pd.DataFrame()
```

```python
for i in range(1,542):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=201dadbac9d90432aa44713682a9eb60&language=en-US&page={}".format(i))
    df = pd.DataFrame(response.json()['results'])
    df = df[['id','title','popularity','overview','vote_average','vote_count']]
    data = data.append(df,ignore_index=True)
    
```
